# Preprocess

!cp /content/data/MITAboveFiveK/raw/Canon_EOS_5D/* /content/data/fivek_dataset

Copy all raw images from different model folders to one folder:

```python
%%bash
RAW_DIR="/content/drive/MyDrive/Colab_Notebooks/CSC2529_Project/data/MITAboveFiveK/raw"
DEST="/content/data/fivek_dataset"

mkdir -p "$DEST"

folders=("$RAW_DIR"/*)
total=${#folders[@]}
count=0

for folder in "${folders[@]}"; do
    count=$((count+1))
    percent=$((100*count/total))

    # print progress bar
    printf "\rProgress: [%-50s] %d%%" "$(printf '#%.0s' $(seq 1 $((percent/2))))" "$percent"

    cp "$folder"/* "$DEST"/
done
```
